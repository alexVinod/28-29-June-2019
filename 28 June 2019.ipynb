```python
def powerN(a,n):
    # return a**n
    r=1
    for i in range(0,n):
        r*=a
    return r
powerN(2,10)

def recursivePowerN(a,n):
    if n == 0:
        return 1
    else:
        return a*recursivePowerN(a,n-1)
recursivePowerN(3,10)
```

```python
# Identify all Primes in a range in Filter

def prime_number(n):
    c=0
    for i in range(1,n+1):
        if (n%i==0):
            c=c+1
    if(c==2):
        return True
    return False

def checkPrime(n):
    if n < 2:
        return False
    for i in range(2,n//2+1):
        if n%i == 0:
            return False
    return True
lb,ub = 500, 600

#li=[1,2,3,4,5,6]
#list(filter(prime_number,li))
primeList=list(filter(checkPrime, range(lb,ub+1)))
primeList2=[ i for i in range(lb,ub+1) if checkPrime(i)]
#primeList3=list(map(i, for i in range(lb,ub+1)))
primeList2
```

### Applying Functional Programming to thr Marks Analysis Application


```python
# Generate Marks data
from random import randint
def generateMarks(n, lb,ub):
    filepath = './DataFiles/marks.txt'
    with open(filepath,'w') as f:
        for i in range(n):
            marks = randint(lb,ub)
            f.write(str(marks)+'\n')
generateMarks(300,0,100)
```

```python
# Marks Analysis
# Class Analysis, 
# % of Passed, Failed and Distinction, 
# Frequency of Highest & Lowest Marks.
import re
def readMarksList(filepath):
    with open(filepath) as f:
        filedata = f.read().split()
        return list(map(int,filedata))
    
def classAverage(filepath):
    markslist = readMarksList(filepath)
    return sum(markslist)//len(markslist)
filepath = './DataFiles/marks.txt'            
classAverage(filepath)
```

```python
def percntageFailed(filepath):
    markslist = readMarksList(filepath)
    failedcount = len(list(filter(lambda mark: mark <40, markslist)))
    # return failedcount
    return (failedcount/len(markslist))*100
percntageFailed(filepath)
```

```python
def percentagePassed(filepath):
    markslist = readMarksList(filepath)
    # passedcount = len(list(filter(lambda mark:mark>=40,markslist)))
    return 100 - percntageFailed(filepath)
percentagePassed(filepath)
```

```python
def percntageDistinction(filepath):
    markslist = readMarksList(filepath)
    distinctioncount = len(list(filter(lambda mark: mark >70, markslist)))
    # return failedcount
    return (distinctioncount/len(markslist))*100
percntageDistinction(filepath)
```

```python
def highestMarkFrequency(filepath):
    markslist = readMarksList(filepath)
    return [markslist.count(max(markslist)),max(markslist)]
highestMarkFrequency(filepath)
```

```python
def lowestMarkFrequency(filepath):
    markslist = readMarksList(filepath)
    return [markslist.count(min(markslist)),min(markslist)]
lowestMarkFrequency(filepath)
```
